```python
information['pricedif'] = information.apply( lambda row : row.close - row.open , axis=1 )
```

```python
#amar tosifi 2 asli


df = information[['Symbol', 'timestamp', 'pricedif']].copy()

df['timestamp'] = pd.to_datetime(df['timestamp'])

unique_symbols = df['Symbol'].unique()

unique_symbols.sort()

result_df = pd.DataFrame(columns=['Symbol 1', 'Symbol 2', 'Count'])

for i in range(len(unique_symbols)):
    for j in range(i+1, len(unique_symbols)):
        sym1 = unique_symbols[i]
        sym2 = unique_symbols[j]

        filtered_df = df.loc[(df['Symbol'] == sym1) | (df['Symbol'] == sym2), ['Symbol', 'timestamp', 'pricedif']]

        grouped_df = filtered_df.groupby('timestamp').agg({'Symbol': list, 'pricedif': lambda x: x.prod()})

        count = (grouped_df['pricedif'] > 0).sum()

        result_df = pd.concat([result_df, pd.DataFrame([[sym1, sym2, count]], columns=['Symbol 1', 'Symbol 2', 'Count'])])

result_df = result_df.sort_values(by='Count', ascending=False)

print(result_df)

```

```python
def ci_exp(sample: np.ndarray, alpha=0.02) :

    n = len(sample)
    # MLE of lambda
    sample_mean = np.mean(sample)

    # Compute the sum of samples
    sample_sum = np.sum(sample)

    # The sample sum is gamma distributed
    gamma_shape = n
    gamma_scale = 1 / sample_mean

    chi2_df = 2 * n

    # Confidence interval for lambda
    lower_bound = (2 * sample_sum) / chi2.ppf(q=(1 - (alpha / 2)), df=chi2_df)
    upper_bound = (2 * sample_sum) / chi2.ppf(q=alpha / 2, df=chi2_df)
    ci = (lower_bound, upper_bound)

    return ci

symbol_list = list(set(df['Symbol']))
m = []
for i in symbol_list :
  info=information[information['Symbol'] == i]['volume'].mean()
  m.append(info)
sample = np.array(m)
ci = ci_exp(sample)
print(ci)
plt.hist(sample)
plt.show()
```
