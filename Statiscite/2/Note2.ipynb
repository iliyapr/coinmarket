```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
cointag = pd.read_csv("coin_tags.csv")
coin = pd.read_csv("coins200.csv")
information = pd.read_csv("coins_information.csv")
tags = pd.read_csv("tags.csv")
```

```python
information['pricedif'] = information.apply( lambda row : row.close - row.open , axis=1 )
result_df = pd.DataFrame(columns=['Symbol 1', 'Symbol 2', 'Count'])
```

```python
df = information[['Symbol', 'timestamp', 'pricedif']].copy()
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
unique_symbols = df['Symbol'].unique()
unique_symbols.sort()
```

```python
for i in range(len(unique_symbols)):
    for j in range(i+1, len(unique_symbols)):
        
        sym1 = unique_symbols[i]
        sym2 = unique_symbols[j]
        
        filtered_df = df.loc[(df['Symbol'] == sym1) | (df['Symbol'] == sym2), ['Symbol', 'timestamp', 'pricedif']]
        
        grouped_df = filtered_df.groupby('timestamp').agg({'Symbol': list, 'pricedif': lambda x: x.prod()})
        
        count = (grouped_df['pricedif'] > 0).sum()
        
        result_df = pd.concat([result_df, pd.DataFrame([[sym1, sym2, count]], columns=['Symbol 1', 'Symbol 2', 'Count'])])
```

```python
result_df = result_df.sort_values(by='Count', ascending=False)
result_df
```

```python
result_df = result_df.sort_values(by='Count', ascending=False)
result_df
```

```python
result_df.reset_index(inplace=False)
result_df = result_df[["Symbol 1", "Symbol 2", "Count"]]
result_df.to_csv('GodDamnIt.csv')
result_df.reset_index(inplace=True)
result_df
```
